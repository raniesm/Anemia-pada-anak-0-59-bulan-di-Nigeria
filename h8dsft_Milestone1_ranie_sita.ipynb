# BAB IV. DATA CLEANING

```python
import pandas as pd

# Load data
data = pd.read_csv("children_anemia.csv")

# Mengganti nilai 'NaN' dengan nilai yang sesuai
data.fillna(value={'Hemoglobin level adjusted for altitude and smoking (g/dl - 1 decimal)' : 0,
                 'Hemoglobin level adjusted for altitude (g/dl - 1 decimal)' : 0,
                 'Anemia level' : 'Unknown',
                 'Anemia.level.1' : 'Unknown',
                 'Currently residing with husband/partner' : 'Unknown',
                 'When child put to breast' : 'Unknown',
                 'Had fever in last two weeks' : 'Unknown',
                 'Taking iron pills, sprinkles or syrup' : 'Unknown'}, inplace=True)

# Menghapus baris dengan nilai null
data_cleaned = data.dropna()

# Mengganti nilai 'Anemia level' menjadi data numerik
anemia_level_mapping = {'Mild' : 1, 'Moderate' : 2, 'Severe' : 3, 'Unknown' : 0}
data_cleaned["Anemia level"] = data_cleaned["Anemia level"].map(anemia_level_mapping)

# Merubah tipe data kolom Births in last five year ke string
data_cleaned["Births in last five years"] = data_cleaned["Births in last five years"].astype(str)

# Memperbaiki nilai yang tidak sesuai pada  kolom Wealth index combined
data_cleaned["Wealth index combined"].replace({"Richer" : "Rich"}, inplace=True)

# Menghapus data duplikat
data_cleaned = data_cleaned.drop_duplicates()

# Merubah semua teks dalam kolom menjadi huruf kecil
data_cleaned["Type of place of residence"] = data_cleaned["Type of place of residence"].str.lower()

# Menyimpan dataframe yang sudah di cleanning ke dalam file csv
data_cleaned.to_csv("children_anemia_cleaned.csv", index=False)

# Menamppilkan informasi data setelah cleaning
print(data_cleaned.info())
print(data_cleaned.head())
```

# BAB V. ANALISIS DAN PERHITUNGAN

###  1. DATA VISUALISASI 1 : Apakah terdapat perbedaan signifikan dalam tingkat anemia anak antara ibu dengan berbagai tingkat pendidikan?


```python
from scipy import stats

# Memisahkan data untuk dua kelompok pendidikan yang ingin dibandingkan
pendidikan_no_education = data_cleaned[data_cleaned['Highest educational level'] == 'No education']['Anemia level']. dropna()
pendidikan_secondary = data_cleaned[data_cleaned['Highest educational level'] == 'Secondary']['Anemia level']. dropna()
pendidikan_primary = data_cleaned[data_cleaned['Highest educational level'] == 'Primary']['Anemia level']. dropna()
pendidikan_higher = data_cleaned[data_cleaned['Highest educational level'] == 'Higher']['Anemia level']. dropna()


# Periksa jumlah samppel untuk masing-masing kelompok pendidikan
print("Jumlah sampel pada pendidikan no education:", pendidikan_no_education.shape[0])
print("Jumlah sampel pada pendidikan secondary:", pendidikan_secondary.shape[0])
print("Jumlah sampel pada pendidikan primary:", pendidikan_primary.shape[0])
print("Jumlah sampel pada pendidikan higher:", pendidikan_higher.shape[0])


# Lakukan uji t-student
f_statistic, p_value = stats.f_oneway(pendidikan_no_education, pendidikan_secondary, pendidikan_primary, pendidikan_higher)

# Menampilkan nilai t statistic dan p-value
print("Nilai f-statistic:", f_statistic)
print("Nilai p-value:", p_value)

# Interpretasi hasil
alpha = 0.05
if p_value < alpha:
    print("Terdapat perbedaan yang signifikan dalam tingkat anemia antara kelompok pendidikan.")
else:
    print("Tidak terdapat perbedaan yang signifikan dalam tingkat anemia antara kelompok pendidikan.")

```

```python
# Digram boxplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('children_anemia_cleaned.csv')

# Menyiapkan data untuk boxplot
data_boxplot = data[data['Highest educational level'].isin(['No education', 'Secondary', 'Primary', 'Higher'])]
data_boxplot = data_boxplot.dropna(subset=['Anemia level', 'Highest educational level'])

# Membuat boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Highest educational level', y='Anemia level', data=data_boxplot, order=['No education', 'Secondary', 'Primary', 'Higher'])

# Memberikan judul dan label
plt.title('Perbandingan Tingkat Anemia berdasarkan Pendidikan Ibu')
plt.xlabel('Pendidikan Ibu')
plt.ylabel('Tingkat Anemia')

# Menampilkan plot
plt.grid(True)
plt.show()
```

- Hasil uji menghasilkan nilai f-statistic sebesar 1.3777831005052525 dan nilai p-value sebesar 0.2475308657412801.
- Dengan menggunakan alpha sebesar 0.05, nilai p-value yang lebih bsar dari alpha menunjukkan tidak ada perbedaan signifikan antara  tingkat anemia pada anak dan pendidikan ibu

Insight dari boxplot:

###  2. DATA VISUALISASI 2 : Bagaimana pola distribusi tingkat anemia pada anak dalam kelompok-kelompok indeks kekayaan rumah tangga yang berbeda?

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("children_anemia_cleaned.csv")

# Mapping kategori kekayaan
wealth_mapping = {
    'Poorest' : ['Poorest'],
    'Poorer' : ['Poorer'],
    'Middle' : ['Middle'],
    'Rich' : ['Rich'],
    'Richest' : ['Richest']
}

# Mengelompokkan data berdasarkan kategori kekayaan
data['Wealth group'] = data ['Wealth index combined'].apply(lambda x: next((key for key, value in wealth_mapping.items() if x in value), None))

# Analisis distribusi tingkat anemia dalam kelompok indeks kekayaan rumah tangga yang berbeda
anemia_distribution = data.groupby("Wealth index combined")["Anemia level"].value_counts().unstack().fillna(0)

# Plotting chart
anemia_distribution.plot(kind='bar', stacked=True)
plt.title('Distribusi Tingkat Anemia pada Anak dalam Kelompok-Kelompok Indeks Kekayaan Rumah Tangga yang Berbeda')
plt.xlabel('Kelompok Indeks Kekayaan Rumah Tangga')
plt.ylabel('Proporsi Tingkat Anemia')
plt.xticks(rotation=45)
plt.legend(title='Anemia level')
plt.tight_layout()
plt.show()
```

Insight :
- Pemetaan tingkat anemia : Data telah dibagi menjadi 5 kelompok indeks kekayaan rumah tangga yaitu: middle, poorer, poorest, rich, dan richest. Tingkat anemia pada anak kemudian dianalisis berdasarkan kelompok indeks kekayaan tersebut, dengan tingkat anemia telah dikategorikan dari 0.0 hingga 3.0 dimana nilai yang lebih tinggi mengindikaskan tingkat anemia yang lebih serius.
- Kelompok 'Poorest' dan 'Poorer' memiliki tingkat anemia yang lebih tinggi dibandingkan dengan kelompok 'Rich' dan 'Richest'. ### 7. Statistik Inferensial: Uji Chi-Square untuk hubungan antara penggunaan kelambu dan tingkat anemia pada anak
### - Apakah penggunaan kelambu berhubungan dengan tingkat anemia pada anak secara signifikan?

```python
# Ho = Tidak ada hubungan antara penggunaan kelambu dan tingkat anemia pada anak
# Ha = Ada hubungan antaran penggunaan kelambu dan tingkat anemia pada anak
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
data = pd.read_csv("children_anemia_cleaned.csv")

# Membuat tabel kontingensi
contingency_table = pd.crosstab(data["Have mosquito bed net for sleeping (from household questionnaire)"], data["Anemia level"])

# Melakukan uji chi-square
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Menampilkan hasil uji
print("Chi-Square Statistic:", chi2)
print("p-value:", p)

# Menafsirkan hasil uji
alpha = 0.05
if p < alpha:
    print("Terdapat hubungan signifikann antara penggunaan kelambu dan tingkat anemia pada anak.")
else:
    print("Tidak terdapat hubungan signifikann antara penggunaan kelambu dan tingkat anemia pada anak.")
```

Penjelasan :
Analisis inferensial menggunakan Chi-Square untuk menila apakah ada hubungan signifikan antara penggunaan kelambu saat tidur dan tingkat anemia pada anak. Berikut adalah hasil insight yang didapatkan dari analisis tersebut :

1. Hipotesis :
    - Ho = Tidak ada hubungan antara penggunaan kelambu dan tingkat anemia pada anak
    - Ha = Ada hubungan antaran penggunaan kelambu dan tingkat anemia pada anak
2. Hasil uji chi-square :
    - Nilai statistik chi-square adala sekitar 6.265
    - Nilai p-value yang diperoleh adalah sekitar 0.0994
3. Penafsiran hasil :
    - Dengan nilai p-value yang lebih besar dari 0.05, hipotesis alternatif ditolak.
4. Kesimpulan :
    - Dari analisis ini, disimpulkan bahwa tidak ada hubungan antara penggunaan kelambu dan tingkat anemia pada anak dalam sampel yang ditelit.
